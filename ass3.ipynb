```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("CAR - CAR (1).csv")
df
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=df,x='year',y='km_driven',hue='owner')
plt.title("Scatterplot")
plt.show()
# Old cars are driven more.

# New cars are driven less.

# Cars with more owners usually have more kilometers driven.
#There is a negative relationship between year and km_driven.
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(data=df.sample(100),x='km_driven',y='selling_price')
plt.show()
# here is a negative relationship between km_driven and selling_price.

# Cars driven more generally have lower selling prices.
```

```python
sns.histplot(df['year'],bins=30,kde=True)
plt.show
# Most cars in the dataset belong to recent years.

# Older model cars are fewer in number.
```

```python
sns.boxplot(data=df,x='fuel',y='year')
plt.show
# Petrol and Diesel cars are available across many manufacturing years.

# Electric vehicles are mostly newer models.
```

```python
sns.jointplot(data=df.sample(100),x='km_driven',y='selling_price')
plt.show()
#Jointplot is used to analyze both relationship and distribution of two numerical variables simultaneously.
#scatterplot shows reltaion between sp and km ,hist shows km and sp ranges
```

```python
sns.barplot(data=df,x='owner',y='selling_price')
plt.show
# > Test drive cars may also have higher selling prices because they are usually newer vehicles with lower usage.
#First-owner cars generally have higher selling prices. Cars with multiple owners tend to have lower selling prices.
```

```python
sns.violinplot(data=df,x='seller_type',y='selling_price')

plt.show()
# Dealer vehicles show a wider selling price distribution.


# Individual sellers mostly sell lower-priced cars.


# Trustmark Dealers generally sell medium to higher-priced vehicles.
```

```python
transmission_c=df['transmission'].value_counts()
plt.pie(transmission_c,labels=transmission_c.index,autopct='%1.1f%%')
plt.show()
```

```python
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show
```

```python
sns.kdeplot(data=df,x='year',y='km_driven')
plt.show()
#KDE plot is used to visualize density distribution of data.
```

```python
plt.scatter(df['fuel'],
            df['year'],
            s=df['year'])

plt.xlabel("KM Driven")
plt.ylabel("Selling Price")
plt.title("Bubble Plot")
plt.show()
```
